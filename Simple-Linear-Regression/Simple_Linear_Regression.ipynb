```python
import pandas as pd

data = {
    "Hours_Studied": [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6,
                      6.5, 7, 7.5, 8, 8.5, 9, 9.5, 10, 10.5, 11],
    "Marks": [32, 41, 44, 48, 57, 59, 61, 67, 66, 73,
              75, 77, 81, 82, 87, 86, 91, 92, 96, 94]
}

df = pd.DataFrame(data)
print(df)
```

```python
df=pd.read_csv('hrs_marks.csv')
print(df.head())
```

```python
plt.scatter(df['Hours_Studied'],df['Marks'])
plt.xlabel('HRS')
plt.ylabel('Marks (numeric)')
plt.show()
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
print(lr.fit(x_train,y_train))
print(x_test)
lr.predict(x_test.iloc[0].values.reshape(1,1))
plt.scatter(df['Hours_Studied'], df['Marks'])
y_pred = lr.predict(x)
```

```python
plt.plot(df['Hours_Studied'], y_pred, color='red')
plt.xlabel('Hours Studied')
plt.ylabel('Marks')
plt.show()
```

```python
m=lr.coef_
b=lr.intercept_
print(m)
print(b)
```
