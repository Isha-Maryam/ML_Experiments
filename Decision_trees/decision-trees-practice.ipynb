```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier 
```

```python
def analyzer(max_depth):
    data = pd.read_csv('/kaggle/input/datasets/d4rklucif3r/social-network-ads/Social_Network_Ads.csv')
    
    print("Your dataset columns are:", list(data.columns))
    
    if 'Age' in data.columns and 'EstimatedSalary' in data.columns:
        X = data[['Age', 'EstimatedSalary']].values
    else:
    
        X = data.iloc[:,].values 
        
    y = data.iloc[:, -1].values 
    
    print(f"Verified X shape: {X.shape}") 
    clf = DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(X, y)
    
    a = np.arange(start=X[:, 0].min() - 1, stop=X[:, 0].max() + 1, step=0.1)
    b = np.arange(start=X[:, 1].min() - 1, stop=X[:, 1].max() + 1, step=100)
    XX, YY = np.meshgrid(a, b)
    
    input_array = np.array([XX.ravel(), YY.ravel()]).T
    labels = clf.predict(input_array)

    plt.contourf(XX, YY, labels.reshape(XX.shape), alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.show()
```

```python
analyzer(max_depth=1) 
```
